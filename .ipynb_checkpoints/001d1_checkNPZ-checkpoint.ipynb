```python
# 检查训练集中的文件是否存在
import os
npz_dir = 'cafa_npz_source/'
with open('cafa_deepfri/MF_train.txt', 'r') as f:
    proteins = [line.strip() for line in f]
missing = [p for p in proteins if not os.path.exists(f'{npz_dir}/{p}.npz')]
print(f'训练集: {len(proteins)} 个蛋白质')
print(f'缺失文件: {len(missing)} 个')
if missing:
    print('前10个缺失的文件:')
    for p in missing[:10]:
        print(f'  {p}.npz')
```

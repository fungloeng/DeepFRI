```python
import tensorflow as tf
import glob

files = glob.glob('./tfrecords/cafa_cc_train*.tfrecords')
if files:
    dataset = tf.data.TFRecordDataset(files[0])
    for serialized in dataset.take(1):
        # 先读取所有特征，看看有哪些键
        example = tf.train.Example()
        example.ParseFromString(serialized.numpy())
        
        # 打印所有特征键
        print("TFRecord中的特征键:")
        for key in example.features.feature:
            feature = example.features.feature[key]
            if key == 'cc_labels':
                # 检查标签维度
                if feature.int64_list.value:
                    print(f"  {key}: 维度 = {len(feature.int64_list.value)}")
                elif feature.bytes_list.value:
                    print(f"  {key}: 是bytes类型")
            else:
                print(f"  {key}: {type(feature)}")
        break
```
